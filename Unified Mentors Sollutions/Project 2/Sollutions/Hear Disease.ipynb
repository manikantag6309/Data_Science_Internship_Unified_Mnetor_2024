```python
import pandas as pd

# Load the dataset
data = pd.read_csv(r"C:\Users\manik\Downloads\Heart Disease data.csv")

# Basic data cleaning 
data = data.dropna()

# Mapping categorical variables for better understanding in Tableau
data['sex'] = data['sex'].map({1: 'Male', 0: 'Female'})
data['cp'] = data['cp'].map({0: 'Typical Angina', 1: 'Atypical Angina', 2: 'Non-anginal Pain', 3: 'Asymptomatic'})
data['thal'] = data['thal'].map({0: 'Normal', 1: 'Fixed Defect', 2: 'Reversible Defect'})
data['target'] = data['target'].map({1: 'Heart Disease', 0: 'No Heart Disease'})
print(data.head())

# Export cleaned data for Tableau
data.to_csv("cleaned_heart_disease_data.csv", index=False)

```

```python
# EDA 
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting Heart Disease by Gender
sns.countplot(x='sex', hue='target', data=data)
plt.title('Heart Disease by Gender')
plt.show()

# Filter numeric columns only
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Plotting Correlation Heatmap
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
